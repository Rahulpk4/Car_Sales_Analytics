```python
df = df.dropna(how="all")
```

```python
df[df.isnull().any(axis=1)]
```

```python
df['Cylinders'].fillna(0, inplace=True)
```

```python
df.head()
```

```python
df['MSRP'] = df['MSRP'].apply(lambda x: x.split('$')[1])
```

```python
df.head()
```

```python
df['Invoice'] = df['Invoice'].apply(lambda x: x.split('$')[1])
```

```python
df.head()
```

```python
df['Invoice'] = df['Invoice'].apply(lambda x: x.replace(" ",""))
df['MSRP'] = df['MSRP'].apply(lambda x: x.replace(" ",""))
df['Invoice'] = df['Invoice'].apply(lambda x: x.replace(",",""))
df['MSRP'] = df['MSRP'].apply(lambda x: x.replace(",",""))
df['MSRP'] = pd.to_numeric(df['MSRP'])
df['Invoice'] = pd.to_numeric(df['Invoice'])
```

```python
df.head()
```

```python
df=df.rename(columns={'MSRP':'MSRP_Dollars', 'Invoice':'Invoice_Dollars'})
```

```python
print(df.Type.unique())
print(df.Origin.unique())
print(df.DriveTrain.unique())
print(df.Cylinders.unique())
```

```python
weight_max = df.groupby('Type')['Weight'].max().reset_index()
weight_avg = df.groupby('Type')['Weight'].mean().reset_index()
```

```python
engine_size_avg = df.groupby('Make')['EngineSize'].mean().reset_index()
```

```python
engine_size_avg=engine_size_avg.sort_values(by="EngineSize", ascending=False).reset_index()
engine_size_avg.drop(columns=["index"])
```

```python
horsepower_avg = df.groupby('Make')['Horsepower'].mean().reset_index()
```

```python
horsepower_avg=horsepower_avg.sort_values(by="Horsepower", ascending=False).reset_index()
horsepower_avg.drop(columns=["index"])
```

```python
df.corr()
```

```python
order_invoice = df.sort_values(by="Invoice_Dollars", ascending=False).reset_index()
```

```python
fig = plt.figure(figsize=(12,6))
plt.scatter(x=order_invoice['Horsepower'], y=order_invoice['Invoice_Dollars'], c="red")
plt.scatter(x=order_invoice['Horsepower'], y=order_invoice['MSRP_Dollars'], c="blue")
plt.xlabel('Horsepower')
plt.ylabel('Dollars')
plt.legend(['Invoice_Dollars', 'MSRP_Dollars'])
plt.title('Horsepower and Invoice/MSRP price correlation')
plt.show()
```

```python
fig = plt.figure(figsize=(12,6))
plt.scatter(x=order_invoice['EngineSize'], y=order_invoice['Cylinders'], c="green")
plt.xlabel('Engine Size')
plt.ylabel('Cylinders')
plt.title('Engine Size and Cylinder capacity correlation')
plt.show()
```
